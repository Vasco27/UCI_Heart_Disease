# Applying Feature Selection
In this section both the complete dataset and the feature selected dataset and presented and compared.

## Complete Dataset
The complete dataset is shown just as described before.

Firstly, a few values are removed for being wrong. The dataset is also shuffled, this is done because it was ordered by the target variable.

```r
uci_heart = read_csv("heart.csv", col_types = c("nffnnffnfdffff"))

#remove wrong thal values
uci_heart = uci_heart[uci_heart$thal != 0, ]

uci_heart$thal = droplevels(uci_heart$thal, exclude = "0")

uci_heart$target = relevel(uci_heart$target, "0")
uci_heart$cp = relevel(uci_heart$cp, "1")

dim(uci_heart)
summary(uci_heart)

set.seed(1)
uci_heart = uci_heart[sample(nrow(uci_heart)), ] #shuffle data, randomize dataframe
```

## Feature Selection Dataset
As explained before, the feature selection was performed using the **lasso** method, in which any variable whose coefficient turned out to be **0** was removed from the model.

The feature selection with the lasso was performed as such, any variable with the **coefficient equal to 0 is removed**. For **numeric variables** the column is simply removed, but for **categorical values** there is a choice, since these types of variables have multiple categories, the lasso coefficients can be zero for one category but significant for another. In this cases, the decision is based on the majority of the categories being significative, for example, the **cp** variable has **two insignificant categories** and **one significant**, therefore this variable is eliminated, while the **ca** variable has **one insignificant category** and **three significant**, therefore this variable is kept. In case of a **tie**, the variable is removed if the significant coefficient is small and the number of rows that have that category is high. 

It is important to remember that, **if we remove one category, we have to remove the remaining ones**. [[2]](https://stats.stackexchange.com/questions/168928/categorical-variables-in-lasso-regression)

As we can see, the feature selection dataset will make the models training a lot **easier to interpret**, due to the fact that the amount of variables is reduced drastically, but it can cause **loss of information**, due to the variables that are not fitting in the model.

```r
df.reduced = uci_heart[, !(names(uci_heart) %in% 
        c("age", "cp", "trestbps", "fbs", "chol"))]

dim(df.reduced)
summary(df.reduced)
```

# Logistic Regression
In this chapter we explore the behaviour of a logistic regression model applying to both datasets (reduced and complete), since the model has no parameters, there will be no experiments with parameter tuning.

## Complete model
Since this is the first model we train, we will start by splitting the data into a train and test set, the train set will contain **70%** of the total data and the test set will contain the remaining **30%**.

The **uci_heart** serves as the **complete** dataset.

In this step, we fit two different models, a complete one and another one that is reduced using the stepwise algorithm.

```r
ret = train_test_split(uci_heart, train_size = 0.7)
train.complete = ret$train; test.complete = ret$test; y_train = train.complete$target; y_test = test.complete$target

set.seed(1)
glm.complete.fit = glm(target ~ ., data = train.complete, family = "binomial")
summary(glm.complete.fit)

glm.completestep.fit = step(glm.complete.fit, trace = 0)
summary(glm.completestep.fit)
```

## Reduced model

In this section, the logistic model is trained using the reduced dataset. The procedure is the same as the one for the complete model, the dataset is divided into a train and test set, two models are trained (reduced and reduced with stepwise selection) and finally, their accuracy is measured. 

```r
ret = train_test_split(df.reduced, 0.7)
train.reduced = ret$train; test.reduced = ret$test;

set.seed(1)
glm.reduced.fit = glm(target ~ ., data = train.reduced, family = "binomial")
summary(glm.reduced.fit)

glm.reducedstep.fit = step(glm.reduced.fit, trace = 0)
summary(glm.reducedstep.fit)
```

## Evaluation and results

Since this is a medical dataset and the variable we are trying to predict tells us if a patient has a heart disease or not, it is very important to predict correctly when the patient **has indeed this disease**, or in other words, when the target variable is **1**. [[3]](https://www.machinelearningplus.com/machine-learning/evaluation-metrics-classification-models-r/)

In light of this characteristic, the most important metric is not the overall accuracy of the model, but the **sensitivity**, the **precision** and the **recall**.
* **Recall** -> It is the same as the **Sensitivity** and corresponds to the percentage of **real** "1s" that were correctly predicted.
* **Precision** -> Corresponds to the percentage of **predicted** "1s" that were correcly predicted.
* **F1 score** -> Combines the precision and the recall in one single metric

### ROC curve and threshold study

Even though correctly classifying "0s" is not as important as correctly classifying "1s", it is still important to try and mitigate the misclassification of these values. Therefore, the **ROC** metric is studied and the **AUC** is compared between models. This metric will allows us to visualize the overall goodness of the model. [[4]](https://www.ritchieng.com/machine-learning-evaluate-classification-model/)

The ROC curves allow us to see the best threshold to use in the model's predictions, in order to have a **high tpr** and an **acceptable fpr**.

```r
models.comp = list(glm.complete.fit, glm.completestep.fit)
models.red = list(glm.reduced.fit, glm.reducedstep.fit)
models.names = c("Complete model", "Complete with stepwise", "Reduced model", "Reduced with stepwise")

plot_ROCCurve(models.comp, models.red, predict_function = make_preds, x.models = models.names)
```

The **AUC** is a generic metric that shows the goodness of a model, in this cases the AUCs are quite high, which means that the model has the potential to be very good. [[4]](https://www.ritchieng.com/machine-learning-evaluate-classification-model/)

If the **threshold** that indicates if a class is classified as positive or negative (currently **0.5**) is **decreased**, then the probabily of predicting a value as positive will increase, i.e. the **sensitivity increases** while the **specifity decreases**, which in this case can be worth, since **predicting the positive class correctly is more important than predicting the negative class correctly**. [[4]](https://www.ritchieng.com/machine-learning-evaluate-classification-model/)

Using these **ROC curves**, a threshold can be found be found, for each model, that **increases the True Positive Rate** while maintaining an **acceptable False Positive Rate**. In this case, we try to find a threshold that provides a tpr higher than 80% and fpr lower than 30%.

```r
metrics = evaluate_models(models.comp, models.red, predict_function = make_preds, x.models = models.names, best.thresholds = TRUE, tpr.threshold = 0.8, fpr.threshold = 0.3, 
                          title = "Different metrics for each model with the Logistic Regression method\n(Using the best thresholds for the recall)")
```

In the plot above, we can see the metrics of the **logistic regression** method while taking advantage of the **best thresholds** to boost the **recall**.

An interesting thing to see in this plot is that the only metric that remains unaltered and similar between models is the **AUC**, due to the fact that the AUC **does not** depend on the threshold for the predictions.

Since the threshold for the **reduced model with stepwise selection** allowed a better fpr (**higher specificity**), the model ends up having a higher overall accuracy than the other. However, this is far from being the best model for this problem, since the most important metric is the **recall**.

```r
metrics = evaluate_models(models.comp, models.red, predict_function = make_preds, x.models = models.names, 
                         title = "Different metrics for each model with the Logistic Regression method\n(Using the standard threshold)")
```

### Comparing thresholds
There are **three** main differences between using the best thresholds instead of the standard 0.5 threshold, in regard to the **Logistic Regression** model:
* The model has a **higher recall* when used with the **best thresholds**.
* Taking into account the example of the **reduced model with stepwise regression**, we can easily tradeoff some recall for **higher accuracy**, however this might not be worth it due to the medical nature of the dataset.
* Overall, using the best threshold leads to **higher accuracy**, in addition to leading to a higher recall, making the study of the best threshold very worthwhile.

### Comparing models
There are **several** differences between the models used:
* In terms of most metrics, the **stepwise selection models** yield worse results than the normal ones.
* The model using **feature selection** yields **slightly worse** results than the complete model, but removes **4** variables from the model.
* Taking the **recall** in consideration, the **complete model** yields the best result, however the **reduced model** only has less **2%** recall.

## Conclusions
* Using the best threshold leads to better results.
* The recall is main metric to improve.
* The stepwise selection yields worse results.
* The complete and feature selected models yield similar results.
* Objectively, the **best model** is the complete model with the 0.43 threshold.
* Subjectively, the **most interpretable model** is the reduced model with the 0.46 threshold.

# Linear Discriminant Analysis (LDA)

Much like the Logistic Regression algorithm, the LDA focuses on estimating the **conditional probability** of the target variable based on the predictors. However, unlike the logistic regression, LDA models the distribution of each predictor for each target variable, applying after this the bayes theorem to estimate $Pr(Y = k|X = x)$, being X the predictors and Y the targets. [5]

There are a few **advantages** of using LDA instead of Logistic Regression:
* Logistic Regression can be unstable for well-seperated classes, unlike the LDA.
* LDA is better when the **number of observations is small** and the predictors distributions tends to be **normal**. (Could be a huge advantage in this problem, since we have a small dataset)
* LDA is more popular when the problem has **more than two target variables**. (Not the case in this problem) [5]

## Complete model

```r
lda.complete.fit = lda(target ~ ., data = train.complete)
lda.complete.fit
```

Not much can be said from just looking at the model, but we can see that 46% of the dataset constains negative cases and 54% contains positive cases.

Another property of the LDA model is that, given the predictors X, the prediction is computed as such $0.012 * X1 + 0.643 * X2 + ...$, if the result is large the case is positive, otherwise it is negative. [6]

## Reduced model

```r
lda.reduced.fit = lda(target ~ ., data = train.reduced)
lda.reduced.fit
```

For now, it is possible to see that the linear discriminant coefficients have altered slightly, due to the reduction made by feature selection.

## Evaluation and results
Just like the previous methods, the evaluation of the **LDA** will begin by studying the ROC curves in order to find the threshold that yields the best results. This threshold will be defined according to the value that **maximizes the tpr**.

Additionally, there will be an evaluation about the **feature selection dataset** and this dataset will be compared to the complete one in order to understand if making the model more interpretable is worth it for the LDA method.

### ROC Curve and threshold study
Observing the ROC Curves shown below, we can find a threshold that maximizes the tpr, without increasing the fpr by too much.

```r
models.comp = list(lda.complete.fit)
models.red = list(lda.reduced.fit)
models.names = c("Complete model", "Reduced model")

plot_ROCCurve(models.comp, models.red, predict_function = lda_preds, x.models = models.names, plots.display = c(1,2), plot.width = 20, plot.height = 7)
```

The **AUC** of both of these models shows that both of them have the potential to be good, so we have to find a good threshold.

Since we want to maximize the tpr, we choose a threshold that has a **tpr higher than 80%**, but does not hurt the fpr too much, after all it is still important to predict negative cases correctly, therefore we do **not** let the **fpr be higher than 40%**.

```r
metrics = evaluate_models(models.comp, models.red, predict_function = lda_preds, x.models = models.names, best.thresholds = TRUE, tpr.threshold = 0.8, fpr.threshold = 0.4, 
                          title = "Different metrics for each model with the LDA method\n(Using the best thresholds for the recall)")
```

The plot above shows that both models have **96% recall**, which is a very high percentage, **almost 100% of positive cases are predicted correctly**. This makes the **LDA** method a very strong one for this dataset.

Looking at the **Specificity**, we can see that choosing a threshold that benefits the recall, can **hurt the negative cases prediction**, however since the accuracy is still fairly high, these models are still acceptable.

Since both models have the same recall percentage, we can conclude that the reduced model has an advantage over the complete one, because it is **more interpretable** and it still maintains the same recall as the complete. 

```r
metrics = evaluate_models(models.comp, models.red, predict_function = lda_preds, x.models = models.names, 
                         title = "Different metrics for each model with the LDA method\n(Using the standard threshold)")
```

### Comparing thresholds
There are several differences between using the best threshold and the standard threshold:
* The **recall in the reduced model is higher** when we use the best threshold.
* Applying the thresholds has a cost in accuracy, meaning we **lose accuracy** when we apply the best threshold.
* The tradeoff for higher recall is lower specificity, meaning that using the best treshold makes it harder to predict negative cases.

### Comparing models
The feature selection has some impacts on the model that differ from the complete:
* In general, using the smaller dataset means **reducing the accuracy** of the model.
* In the case of LDA, the smaller dataset can compete with the complete one (using the best threshold).

## Conclusions
* Using the best thresholds yields better results.
* The best threshold makes the **reduced model viable**.
* Using LDA, the best model would be the **reduced dataset with the best threshold**.

# Quadratic Discriminant Analysis (QDA)
Just like LDA, the QDA assumes that the observations for each class come from a **gaussian distribution** and applies the Bayes theorem to make predictions. However, **unlike the LDA** which assumes that all classes share the same covariance matrix, this method assumes that each class has its own covariance matrix. [7]

So what is the difference between using LDA and QDA? The answer lies in the bias-variance trade-off:
* By assuming that all classes share a covariance matrix, the LDA becomes less flexible and has lower variance.
* By assuming each class has a different covariance matrix, the QDA becomes computationally heavier.
* LDA tends to do better when we have a small ammount of observations, while QDA gets better results with large datasets (variance has less impact). [7]

## Complete model

```r
qda.complete.fit = qda(target ~ ., data = train.complete)
qda.complete.fit
```

The approach is very similar to LDA. 

The coefficients do not show, because QDA involves a quadratic function of the predictors, rather than the linear one from LDA.

## Reduced model

```r
qda.reduced.fit = qda(target ~ ., data = train.reduced)
qda.reduced.fit
```

Much like the previous methods, the evaluation for the **QDA** will follow the same guidelines. Firstly, we start by studying the ROC curves to look for a good threshold for the predictions that will allow us to maximize the recall.

Secondly, we will study the metrics for both, the predictions with the best threshold and the predictions with the standard threshold.

Finally, a comparison will be make between both models (feature selection and complete) and both thresholds (best and standard) from which we will draw our conclusions.

### ROC Curve and threshold study
We plot the ROC Curves to find a treshold that **maximizes the tpr** without lowering the fpr by too much.

```r
models.comp = list(qda.complete.fit)
models.red = list(qda.reduced.fit)
models.names = c("Complete model", "Reduced model")

plot_ROCCurve(models.comp, models.red, predict_function = lda_preds, x.models = models.names, plots.display = c(1,2), plot.width = 20, plot.height = 7)
```

Observing the plot above, we can see that the **AUC** for both models is **fairly high** but it is the lowest we have seen so far, meaning finding a good threshold might be a challenge for the **QDA** method.

The complete model has good thresholds right after the **20% fpr mark**, while the reduced model has a more linear curve that plateaus around the **15% fpr mark**. The thresholds are chosen taking this information into account.

As per usual, we decide that the tpr has to be **higher than 80%**, while the fpr needs to be **below 30%** due to the complete model ROC Curve.

```r
metrics = evaluate_models(models.comp, models.red, predict_function = lda_preds, x.models = models.names, best.thresholds = TRUE, tpr.threshold = 0.8, fpr.threshold = 0.3, 
                          title = "Different metrics for each model with the QDA method\n(Using the best thresholds for the recall)")
```

As it was expected, because of how the ROC Curve of the complete model shoots up after the 0.2 fpr mark, the **recall for the complete model is very high**.

The reduced model had a much more linear curve, meaning it was harder to find a threshold that maximized the recall, therefore the **reduced model has a much lower recall** than the complete model, as well as a lower accuracy, due to the reduced model **inherently having worse prediction accuracy** than the complete.

In this case, the interpretability-accuracy trade-off between this models might not be worth the risk.

# K-Nearest Neighbors (KNN)
The K-Nearest Neighbors algorithm is a very well known algorithm used for classification problems.

The **main idea** of the algorithm is to classify an observation based on the **K** observations that are **nearest to it**. So, for example if K = 5 and the five nearest neighbors of a certain observation are 3 positives and 2 negatives then the observation will be classified as positive. Of course, it has to use the bayes rule to classify an observation given these near observations. [8].

There are several **advantages** for using the KNN algorithm:
* The KNN makes little assumptions about the data (does not need to be normal, for example).
* Only one parameter to tune (**K**).
* Tuning the parameter K makes it easier to find a good model.

## Tuning the parameter K
In this section, we explore the evolution of different metrics as the parameter K increases. The main metrics we watch out for are the **recall** (because predicting positive cases is crucial) and the **accuracy**, because we still need to maintain a decent accuracy even if the recall is very high.

We need to pay attention for **two things** when studying the behaviour of the K parameter:
* When the parameter **K is low**, there is a high chance that the train error is lower, meaning the algorithm **overfits** the data.
* When the parameter **K is high**, the train error is higher and the algorithm will **underfit** the data.

Taking this into account, we try to choose K values that have a good balance between underfitting and overffiting.

```r
k = c(1,3,5,7,9,11,13,15,19,23,27,31,35,39,43)

traintest_errs = evaluate_knn_traintest_errors(k)
```

In the plot above, we can see how the train and test errors evolve according to the increase in the parameter K for each model (feature selection and complete).

For the **complete model** we can see the expected behaviour of the train and test errors of the KNN method. As K increases the train error increases and the test error decreases until they reach a semi-plateau at around 30% error. So for the complete model, **K > 7** would be the ideal balance to prevent overfitting, since it is when the error starts stabilizing.

For the **reduced model** the behaviour is a little strange, as K increases both the train and test error increase until they plateau between 25% and 30%. For this reduced model, a **K > 13** would be ideal to prevent overfitting, since the **train error is higher than 20%** and it reachs the plateau.

In the next part, we will try to make use of this conclusions to choose the best parameter K for each model, one that will prevent the overfitting of the data and still be able to have good metrics.

```r
ret = evaluate_knn(ks = k)
```

In the plot above, the metrics results for each model (complete and feature selection) are displayed. Interestingly, **both models behave differently** in terms of how metrics change as K increases.

For the **complete model**, we can see that all the metrics behave more or less the same way. From K = 1 until K = 23 the metrics increase and it decreases after that, therefore the model **peaks at K = 23**, which is great because as we have seen before the complete model has less chances of overfitting when K > 7.

The **reduced model** has a completely different behaviour, as we can see the metrics **steadily decline** since K = 1, however the **recall peaks at K = 5** and since the recall is the metric we are trying to maximize we chose that K as the best one for this model. Unfortunately, K = 5 is **lower than 13**, meaning there is a risk of some **overfitting**.

The next part shows a comparison of the metrics between the two KNN models with the best K parameters.

```r
best.metrics.melt = visualize_best_k(ret, best.ks = c(23, 5), x.models = c("Complete model", "Reduced model"))
```

In the last part, we tuned the parameter K in order to find the best KNN model. For the **complete model** the best parameter was **K = 23** and for the **reduced model** the best parameter was **K = 5**.

The metrics are calculated using a test set not previously known by the model.

In the case of **KNN**, the reduced model has better metrics than the complete one, including the recall and the accuracy, which are the most important metrics for this dataset. So for the KNN method, the reduced dataset has the best of both worlds, more interpretability and better metric results.

### Conclusions
* Tuning the K parameter is very important to improve the model.
* Having a good train error is crucial to prevent the overfitting of the model.
* The reduced model is far superior to the complete one, better interpretability and better metrics (especially recall).

# Final conclusions
Finally, we reached the final conclusions. In this section we will take the best models found for each algorithm and compare them, in order to understand which model is the best for this specific dataset.

There are **two** main comparison we need to make, one about the **feature selection vs complete dataset** and another about the **difference between algorithms**.

* The QDA and the Logistic regression were a slight worse fit than the LDA and KNN, specially for the feature selection dataset, making them acceptable while using the complete dataset, but clearly not good fits for this problem.
* Overall, the **LDA was the best fit for this problem**, not only were we able to obtain a high recall (96%) but that recall was also obtain for the feature selection dataset, making it a strong, reliable and interpretable model, almost perfect for predicting medical cases.
* If we do not want to choose the LDA algorithm, we can choose the KNN which makes less assumptions, but obtained some good results as well. For the recall metric in the feature selection dataset the KNN obtain a result of (91%), meaning the model is very interpretable and makes very accurate predictions for positive cases.
